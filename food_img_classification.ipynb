## set dataloader

```python
# get img size

from PIL import Image

def get_img_size(imgPath):
    """
    이미지의 크기를 반환하는 함수

    Args: imgPath (str) : 이미지 파일 경로
    Return: tuple - (너비, 높이)
    
    """

    try:
        with Image.open(imgPath) as img:
            width, height = img.size
            return width, height
    except FileNotFoundError:
        print(f"Error: 이미지 파일을 찾을 수 없습니다. ({imgPath})")
```

```python
#get_img_size()

#basePath
print(orgPath)

# Food Classification dataset/apple_pie/
apple_pie_dir = os.path.join(orgPath, "apple_pie")

# apple_pie 안에 모든 이미지 파일 사이즈 확인
for filename in os.listdir(apple_pie_dir): # 경로명 지정
    if filename.endswith('.jpg'): # 파일 확장자 확인
        file_path = os.path.join(apple_pie_dir, filename)

        # 이미지 사이즈 확인
        width, height = get_img_size(file_path)

        if width and height:
            print(f"파일: {filename}, 너비: {width}, 높이: {height}")
```
